딕셔너리 생성

```python
from typing import Any

member: dict[str, Any] = { 
    "id":"python", 
    "name":"파이썬", 
    "age":20, 
    "tel":"0101231234" 
}
```

요소 찾기

```python
name = member["name"]
print(name)

# job = member["job"]   # KeyError
job = member.get("job")
print(job)
```

```python
member = {
    "name": "홍길동",
    "grade": 3,
    "adult": False,
    "hobby": ["영화", "축구"],
    "school": ("대치초등학교", "서초중학교", "강남고등학교"),
    "family": {"father": "홍삼원", "mother": "김순희"}
}

# 멤버가 졸업한 중학교
middle_school = member["school"][1]
print(middle_school)

# 멤버의 엄마의 이름은?
mother_name = member["family"].get("mother")
print(mother_name)
```

```python
for key in member:
    value = member[key]
    print(f"{key}: {value}")
```

```python
for value in member.values():
    print(value)
```

```python
for key, value in member.items():
    print(f"{key}: {value}")
```

```python
board = {
    "title": "파이썬 도서",
    "writer": "신용권",
    "date": "20180731",
    "content": "AI 엔지니어를 위한 파이썬 언어",
    "hitcount": 10000
}

board["title"] = "파이썬 프로그래밍"
print(board)
```

요소 추가

```python
board = {
    "title": "파이썬 도서",
    "writer": "신용권",
    "date": "20180731",
    "content": "AI 엔지니어를 위한 파이썬 언어",
    "hitcount": 10000
}

board["press"] = "한빛미디어"
print(board)
```

요소 제거

```python
board = {
    "title": "파이썬 도서",
    "writer": "신용권",
    "date": "20180731",
    "content": "AI 엔지니어를 위한 파이썬 언어",
    "hitcount": 10000
}

del board["hitcount"]
print(board)
```

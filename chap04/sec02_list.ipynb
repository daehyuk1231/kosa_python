```python
# 인덱싱으로 변경: 값만 제공
# 슬라이싱으로 변경: 리스트형태로 변경 값을 제공
scores = [99, 95, 93, 91, 88, 84, 78, 75, 73, 70]

# 인덱싱
scores[1] = 100

# 슬라이싱
scores[-3:] = [0, 0, 0]
scores[1:2] = [100]
print(scores)
```

```python
scores = []

scores.append(50)
scores.append(70)
print(scores)

scores.insert(1, 60)
print(scores)
```

```python
# 슬라이싱으로 제거
scores = [99, 95, 93, 91, 88, 84, 99, 75, 73, 70]

# 슬라이싱을 이용해서 끝 3개의 요소를 제거
scores[-3:] = []
print(scores)

# 슬라이싱을 이용해서 1 인덱스의 요소를 제거
scores[1:2] = []
print(scores)

# 인덱싱을 이용해서 제거
del scores[1]
print(scores)

# 요소 값을 직접 제공해서 제거(첫번째 요소만 제거)
# scores.remove(99)
# print(scores)

# 요소 값을 직접 제공해서 모든 요소 제거
while 99 in scores:
    scores.remove(99)
print(scores)

# 모든 요소를 제거
scores.clear()
print(scores)
```

```python
list1 = [10, 20, 30]
list2 = [40, 50, 60, 70, 80]
scores = [list1, list2]
print(scores)

scores = [
    [10, 20, 30], 
    [40, 50, 60, 70, 80]
]
print(scores)

sum: int = 0
count: int = 0
for i in range(len(scores)):
    for k in range(len(scores[i])):
        count += 1
        sum += scores[i][k]
avg: float = sum / count
print(avg)
```

set 생성

```python
# 값의 목록으로 셋 생성
nums: set[int] = {1, 2, 3, 4, 5}
print(nums)

# 리스트로 셋 생성
names = ["홍길동", "홍길서", "홍길남"]
names = set(names)
print(type(names), names)

# 빈요소로 셋을 만들 수 없다.
# cities = {}
# print(type(cities)) # dict 타입
cities = set()
print(type(cities))
```

요소 얻기

```python
chars = { "a", "b", "c", "d", "e" }

while chars:
    value = chars.pop()
    if value == "d" or value == "e":
        print(value)
```

요소 추가

```python
pocket = set()

pocket.add("구슬1")
pocket.add("구슬2")
print(pocket)

pocket.update(["구슬3", "구슬4"])
print(pocket)
```

요소 제거

```python
basket= {'사과', '망고', '오렌지', '수박', '딸기', '포도'}

basket.remove('오렌지')
print(basket)

basket.clear()
```

set 연산

```python
set1 = {2, 4, 6, 8, 10}
set2 = {4, 8, 12, 16}
set3 = {1, 7}
set4 = {2, 4, 6}
```

```python
# 합집합
result = set1.union(set2)
print(result)

result = set1 | set2
print(result)
```

```python
# 교집합
result = set1.intersection(set2)
print(result)

result = set1 & set2
print(result)
```

```python
# 차집합
result = set1.difference(set2)
print(result)

result = set1 - set2
print(result)
```

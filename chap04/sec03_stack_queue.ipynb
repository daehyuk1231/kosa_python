```python
from collections import deque

# 큐 객체 생성
queue = deque() #FIFO

# 메시지 저장하기
queue.append({"from":"홍길동", "to":"고광우", "content":"수업받지 말고 놀자."})
queue.append({"from":"홍길동", "to":"천승현", "content":"좀 놀자"})
queue.append({"from":"홍길동", "to":"김정원", "content":"열심히 하자."})

# 메시지 처리하기
while queue:
    print(queue.popleft())
```

```python
class Member:
    def __init__(self, name:str, ssn:str) -> None:
        self.__name = name
        self.__ssn = ssn
        
    def get_name(self):
        return self.__name
    
    def get_ssn(self):
        return self.__ssn
```

```python
member = Member("홍길동", "123456-1234567")
# print(member.__name)  # AttributeError
# print(member.__ssn)   # AttributeError
print(member._Member__name)     # Name mangling
print(member._Member__ssn)     # Name mangling
print(member.get_name())
print(member.get_ssn())
```

```python
class Member:
    # 초기화 메소드
    def __init__(self, name:str, ssn:str) -> None:
        self.__name = name
        self.__ssn = ssn
        
    # 읽기 속성 정의
    @property
    def name(self):
        return self.__name
    
    # 쓰기 속성 정의
    @name.setter
    def name(self, name:str):
        self.__name = name
    
    # 읽기 속성 정의    
    @property
    def ssn(self):
        return self.__ssn
    
    # 쓰기 속성 정의
    @ssn.setter
    def ssn(self, ssn:str):
        if len(ssn) != 14:
            raise ValueError("주민번호 자리수가 맞지 않음")
        self.__ssn = ssn
```

```python
member = Member("홍길동", "123456-1234567")

# 읽기 속성 사용
print(member.name)
# 쓰기 속성 사용
member.name = "홍길남"
print(member.name)

# 읽기 속성 사용
print(member.ssn)
# 쓰기 속성 사용
member.ssn = "654321-7777777"   #raise ValueError("주민번호 자리수가 맞지 않음")
print(member.ssn)
```

```python
from dataclasses import dataclass


@dataclass
class Member:
    mid: str
    mname: str
    mage: int
    
m1 = Member("user1", "홍길동", 30)
print(m1)
m1
```

```python
m2 = Member("user1", "홍길동", 30)

print(m1 == m2) # 동등 비교
print(m1 is m2) # 번지 비교
```

```python
print(m1.mid)
print(m1.mname)
print(m1.mage)
```

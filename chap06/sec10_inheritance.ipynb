```python
class Person:
    def __init__(self, name:str) -> None:
        self.name:str = name

p1 = Person("홍길동")
print(p1.name)
```

```python
class Student(Person):
    def __init__(self, name: str, sno: str) -> None:
        super().__init__(name)
        self.sno =sno
        
s1 = Student("홍길동", "20260103")
print(s1.name)
print(s1.sno)
```

```python
class A:
    def method1(self):
        print("A.method() 실행")
        
class B:
    def method2(self):
        print("B.method() 실행")

class C:
    def method3(self):
        print("C.method() 실행")        
        
class D(A, B, C):
    pass

d = D()
d.method1()
d.method2()
d.method3()
```

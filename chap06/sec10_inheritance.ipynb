```python
class Person:
    def __init__(self, name:str) -> None:
        self.name:str = name

p1 = Person("홍길동")
print(p1.name)
```

```python
class Student(Person):
    def __init__(self, name: str, sno: str) -> None:
        super().__init__(name)
        self.sno =sno
        
s1 = Student("홍길동", "20260103")
print(s1.name)
print(s1.sno)
```

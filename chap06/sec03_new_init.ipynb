```python
from typing import override, Self

# Car 클래스 정의
class Car:
    # 객체를 생성하는 메소드
    # 클래스 메소드(정적 메소드)
    @override
    def __new__(cls) -> Self:
        print("__new__() 메소드 실행")
        return super().__new__(cls)
    
    # 객체를 초기화하는 메소드
    # 인스턴스 메소드
    @override
    def __init__(self) -> None:
        print("__init__() 메소드 실행")
```

```python
my_car = Car()
```

```python
class Car:
    @override
    def __new__(cls, name, year) -> Self:
        # 생성자 매개값 타입 검사
        if not isinstance(name, str):
            raise TypeError("name은 str 타입이어야 합니다.")
        if not isinstance(year, int):
            raise TypeError("year는 int 타입이어야 합니다.")
        # 매개변수가 유효할 경우에만 객체 생성
        return super().__new__(cls)
    
    @override
    def __init__(self, name, year) -> None:
        pass
```

```python
my_car = Car("그랜저", 2026)
# my_car = Car(2026, "그랜저")
my_car = Car("G80", 2026)
```

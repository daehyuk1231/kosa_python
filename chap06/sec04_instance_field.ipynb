```python
class Member:
    def __init__(self, mid:str, age:int, hobby:list=[]) -> None:
        self.mid = mid
        self.age = age
        self.hobby = hobby
```

인스턴스 필드값 읽기

```python
m1 = Member("홍길동", 30, ["영화", "음악"])
m1.name = "홍길서"
m1.age = 25
m1.hobby = ["게임", "투자"]

m1.hobby.append("수영")
print(m1.hobby)

# 파이썬은 동적 필드(속성)을 지원하지 않음
# m1.job = "개발자"
```

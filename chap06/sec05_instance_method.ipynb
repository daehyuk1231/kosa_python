```python
class Car:
    def __init__(self, speed:int = 0, tire:str|None = None) -> None:
        self.speed = speed
        self.tire = tire
        
    # 인스턴스 메소드를 정의
    def run(self) -> int:
        self.set_speed(60)
        print(f"자동차가 {self.speed}km/h로 달립니다.")
        return self.speed
        
    def set_speed(self, speed:int) -> None:
        self.speed = speed
        
    def change_tire(self, tire:str) -> str:
        self.tire = tire
        return (f"{self.tire}로 교체합니다.")
```

```python
my_car = Car()

current_speed: int = my_car.run()
my_car.set_speed(100)
print(my_car.speed)
print(my_car.change_tire("넥센"))
```

동적 필드 및 동적 메소드 지원

```python
my_car = Car()

# 동적 필드 추가
my_car.company = "현대자동차"
print(my_car.company)

# 동적 메소드 추가
def stop(self):
    print("자동차를 멈춥니다.")
    
import types
my_car.stop = types.MethodType(stop, my_car)
my_car.stop()
```

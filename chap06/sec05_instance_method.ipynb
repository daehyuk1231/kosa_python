```python
class Car:
    def __init__(self, speed:int = 0, tire:str|None = None) -> None:
        self.speed = speed
        self.tire = tire
        
    # 인스턴스 메소드를 정의
    def run(self) -> int:
        self.set_speed(60)
        print(f"자동차가 {self.speed}km/h로 달립니다.")
        return self.speed
        
    def set_speed(self, speed:int) -> None:
        self.speed = speed
        
    def change_tire(self, tire:str) -> str:
        self.tire = tire
        return (f"{self.tire}로 교체합니다.")
```

```python
my_car = Car()

current_speed: int = my_car.run()
my_car.set_speed(100)
print(my_car.speed)
print(my_car.change_tire("넥센 타이어"))
```

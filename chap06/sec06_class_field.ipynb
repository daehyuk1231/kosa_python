```python
from typing import Final


class Car:
    # 클래스 필드
    kind:str = "승용차"
    COMPANY: Final[str] = "현대자동차"
    
    # 객체 초기화
    def __init__(self) -> None:
        # 인스턴스 필드
        self.speed = 0
        
    def run(self) -> None:
        self.speed = 60
        print(f"{Car.COMPANY}가 {self.speed}로 달립니다.")
```

```python
print(Car.kind)
Car.kind = "SUV"
print(Car.kind)

print(Car.COMPANY)
# Car.COMPANY = "기아자동차"
# print(Car.COMPANY)

my_car = Car()
my_car.run()
```

```python
from typing import override


class Car:
    def __init__(self, model, color) -> None:
        self.model = model
        self.color = color
        
    @override
    def __str__(self) -> str:
        return f"__str__:{self.model}-{self.color}"
    
    @override
    def __repr__(self) -> str:
        return f"__repr__:{self.model}-{self.color}"
        
my_car = Car("포르쉐", "흰색")

print(my_car)   # __str__()의 반환값 출력

my_car          # __repr__()의 반환값 출력
```

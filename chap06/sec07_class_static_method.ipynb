```python
class Car:
    # 클래스 필드
    kind = "승용차"
    
    @classmethod
    def method1(cls):
        print("method1() 실행")
        print(Car.kind, cls.kind)
    
    @staticmethod
    def method2():
        print("method2() 실행")
        print(Car.kind)
        
    def method3(self):
        print("method3() 실행")
```

```python
Car.method1()
Car.method2()
car = Car()
car.method3()
# Car.method3()
```

싱글톤 구현하기

```python
from typing import Self


class MemberService:
    # 클래스 변수
    _member_service = None
    
    def __new__(cls) -> Self:
        if cls._member_service == None:
            cls._member_service = cls()
        return cls._member_service
```

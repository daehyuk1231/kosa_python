```python
class Car:
    # 클래스 필드
    kind = "승용차"
    
    @classmethod
    def method1(cls):
        print("method1() 실행")
        print(Car.kind, cls.kind)
    
    @staticmethod
    def method2():
        print("method2() 실행")
        print(Car.kind)
        
    def method3(self):
        print("method3() 실행")
```

```python
Car.method1()
Car.method2()
car = Car()
car.method3()
# Car.method3()
```

싱글톤 구현하기

```python
from typing import Self


class MemberService:
    # 클래스 변수
    _member_service = None
    _initialized = False
    
    # 객체 생성 역할
    def __new__(cls) -> Self:
        if cls._member_service == None:
            # 객체를 최초로 생성할 경우에는 super().__new__(cls) 사용
            cls._member_service = super().__new__(cls)
        return cls._member_service
    
    # 객체 초기화 역할
    def __init__(self) -> None:
        if self._initialized == False:
            print("__init__() 실행해서 초기화됨")
            self.field1 = ""
            self.field2 = ""
            self._initialized = True
            
    # 클래스 메소드를 이용해서 싱글톤 객체 반환
    @classmethod
    def get_instance(cls):
        return cls()
```

```python
member_service_1 = MemberService()
member_service_2 = MemberService()

print(member_service_1 is member_service_2)
```

```python
member_service_3 = MemberService.get_instance()
member_service_4 = MemberService.get_instance()

print(member_service_3 is member_service_4)     # 번지 비교
```

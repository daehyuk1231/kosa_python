```python
class BankError(Exception):
    def __init__(self, message:str = "", error_code="") -> None:
        super().__init__(message)
        self.error_code = error_code
        
# 예외 발생 코드
# raise BankError()
# raise BankError("잔고부족")

class Bank:
    def __init__(self) -> None:
        self.balance: int = 0
    
    def withdraw(self, amount:int):
        if amount > self.balance:
            raise BankError("잔고부족", "E001")
        self.balance -= amount
```

```python
try:
    bank = Bank()

    bank.balance = 10000
    bank.withdraw(1000)
    print(bank.balance)

    bank.withdraw(20000)
except BankError as e:
    message = str(e)
    if e.error_code == "E001":
        print(message, "예금액이 부족합니다.")
    elif e.error_code == "E002":
        print(message, "비밀번호가 틀립니다.")
```

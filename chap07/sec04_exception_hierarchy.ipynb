```python
while True:
    try:
        print("숫자를 입력하세요")
        x = input()
        result = 10 / int(x)
    except ValueError:
        print("숫자가 아닌 수를 입력했습니다.")
    except Exception:
        print("0으로 나눌 수 없습니다.")
    else:
        print("결과:", result)
    finally:
        print("계속 여부 (y|n):")
        exit = input()
        if exit == "n":
            break
```

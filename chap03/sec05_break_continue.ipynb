```python
# 1~10 범위를 반복할때 5가되면 반복을 중지
# while문에서 break 사용
i = 1
while i <= 10:
    if i == 5:
        print("종료")
        break
    print(i)
    i+=1
```

```python
# 1~10 범위를 반복할때 5가되면 반복을 중지
# for 문에서 break 사용

for i in range(1,11):
    if i == 5: 
        print("종료")
        break
    print(i)
```

```python
# 1~10 범위를 반복하면서 3의 배수의 합을 출력하세요.
# while 문에서 continue 사용

i = 1
sum = 0
while i <= 10:
    if i % 3 != 0:
        i+=1
        continue
    sum += i
    i+=1
print(sum)
```

```python
# 1~10 범위를 반복하면서 3의 배수의 합을 출력하세요.
# for 문에서 continue 사용

i=1
sum=0
for i in range(1,11):
    if i % 3 != 0:
        continue
    sum += i
print(sum)
```

```python
available_coffee: int = 5

while available_coffee:
    print("커피를 제공합니다.")
    available_coffee -= 1
    print(f"남은 커피 수: {available_coffee}")
    
print("프로그램 종료")
```

```python
stop: bool = False

while not stop:
    print("입력: ", end="")
    keyboard_input = input()
    if keyboard_input == "음악":
        print("음악을 들려 드릴게요.")
    elif keyboard_input == "영화":
        print("영화를 보여 드릴게요.")
    elif keyboard_input == "중지":
        stop = True
        
print("프로그램 종료")
```

# 문자열 내부에 특수기호를 넣는 방법

- 큰따옴표 넣기

```python
var1 = "지금학습하고 있는 언어는 \"파이썬\"입니다."
print(var1)
```

- 문자열의 길이

```python
# 키보드로 문자열 입력
data = input()
print(data)
print("입력된 문자수(문자열의 길이): ", len(data))
```

- 문자열 인덱싱과 슬라이싱

```python
text = "파이썬 프로그래밍"

# 인덱싱
print(text[0])
print(text[4])

# 슬라이싱
print(text[0:3])
print(text[:3])
```

```python
text = "파이썬 프로그래밍"

# 인덱싱
print(text[-1])

# 슬라이싱
print(text[4:])
print(text[-5:])
```

- 건너뛰기(스텝)

```python
# 슬라이싱 문법: [시작인덱스:끝인덱스:스텝]
numbers = "123456789"
print(numbers[0:7:2])
print(numbers[0::2])
```

```python
# 주민번호에서 생년월일과 뒷자리 7자리를 따로 출력하시오
# 남자인지 여자인지 출력하시오

ssn = "123456-1234567"
first = ssn[:6]
sex = ssn[7]
second = ssn[7:]

print("생년월일:", first)
print(f"뒷자리숫자: {second}")

if sex == "1" or sex == "3":
    print("남자")
else:
    print("여자")
```

- 문자열 결합

```python
str1 = "파이썬"
str1 += " 프로그래밍"
print(str1)
```

```python
str2 = "-" * 50
print(str2)
print("메뉴1 | 메뉴2 | 메뉴3")
print(str2)
```

```python
str3 = "문자열을" " 연결합니다."
print(str3)

str3 = "문자열을" + \
       " 연결합니다."
print(str3)

str3 = ("문자열을" +
       " 연결합니다.")
print(str3)
```

- 문자열 비교

```python
var1 = "Python"
var2 = var1 + "3" # 변수 연산의 결과는 새로운 문자열 객체가 생성됨
var3 = var1 + "3" # 변수 연산의 결과는 새로운 문자열 객체가 생성됨

print(var2 == var3) # 동등 비교(값 비교)
print(var2 is var3) # 번지 비교
```

- 문자열 포함 여부

```python
var = "Python" in "This is Python Book"
print(var)

var = "Python" not in "This is Python Book"
print(var)
```

```python
var = "This is Python Book".lower()
print(var)

var = "python".lower() in "This is Python Book".lower()
print(var)
```

- 문자열 검사

```python
str1 = "abc12345"
print(str1.isalnum())
print(str1.isalpha())
```

```python
str2 = "3"
str3 = "3.5"

print(str2.isdecimal()) # 0, 1, 2, 4, 5, 6, 7, 8, 9
print(str3.isdecimal()) # 0, 1, 2, 4, 5, 6, 7, 8, 9

print(int(str2)) # 문자열을 정수로 변환
# print(int(str3)) # 정수로 변환이 안됨
print(float(str3)) # 문자열을 실수로 변환
```

- 문자열 공백 제거

- 문자열 찾기

```python
str = "123456-3890123"

print(str.index("-"))
print(str.find("-", 7, 9))
```

```python
str = "123456-3890123"

seperater_index = str.find("-")
if seperater_index != -1:
    # 있다면
    part1 = str[:seperater_index]
    part2 = str[seperater_index+1:]
    print(part1)
    print(part2)
```

- 토큰 분리

```python
tel_num = "010-1234-5678"
parts = tel_num.split("-")
print(parts)
print(parts[0])
print(parts[1])
print(parts[2])
```

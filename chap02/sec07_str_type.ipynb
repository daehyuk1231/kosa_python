# 문자열 내부에 특수기호를 넣는 방법

- 큰따옴표 넣기

```python
var1 = "지금학습하고 있는 언어는 \"파이썬\"입니다."
print(var1)
```

```python
# """: doc string이며 내부의 문자열이 그대로 값으로 인식
var2 = """지금학습하고 있는 언어는 "파이썬"입니다."""
print(var1)

var2 = """지금학습하고 있는 언어는
          "파이썬"입니다."""
print(var2)
```

- 문자열의 길이

```python
# 키보드로 문자열 입력
data = input()
print(data)
print("입력된 문자수(문자열의 길이): ", len(data))
```

- 문자열 인덱싱과 슬라이싱

```python
text = "파이썬 프로그래밍"

# 인덱싱
print(text[0])
print(text[4])

# 슬라이싱
print(text[0:3])
print(text[:3])
```

```python
text = "파이썬 프로그래밍"

# 인덱싱
print(text[-1])

# 슬라이싱
print(text[4:])
print(text[-5:])
```

- 건너뛰기(스텝)

```python
# 슬라이싱 문법: [시작인덱스:끝인덱스:스텝]
numbers = "123456789"
print(numbers[0:7:2])
print(numbers[0::2])
```

```python
# 주민번호에서 생년월일과 뒷자리 7자리를 따로 출력하시오
# 남자인지 여자인지 출력하시오

ssn = "123456-1234567"
first = ssn[:6]
sex = ssn[7]
second = ssn[7:]

print("생년월일:", first)
print(f"뒷자리숫자: {second}")

if sex == "1" or sex == "3":
    print("남자")
else:
    print("여자")
```

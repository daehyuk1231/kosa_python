```python
from datetime import datetime

now = datetime.now()

print(now, type(now))
print(now.year)
print(now.month)
print(now.day)
print(now.hour)
print(now.minute)
print(now.second)
print(f"{now.year}.{now.month}.{now.day}")
```

```python
print(now.strftime("%Y.%m.%d"))
print(now.strftime("%Y.%m.%d %H:%M:%S"))
```

```python
# datetime 생성자를 이용해서 지정한 날짜와 시간을 갖는 datetime 객체 생성
target_datetime = datetime(2050, 5, 11, 9, 30, 10)
print(target_datetime)
```

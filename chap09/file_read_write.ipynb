```python
# 온도 센서에서 측정된 데이터를 JSON 문자열로 반환하는 함수
from datetime import datetime
import json
import random


def temperature_sensor():
    # data = {}
    # data["sensor"] = "온도"
    # data["value"] = random.randint(20, 30)
    # data["datetime"] = datetime.now().strftime("%Y.%m.%d %H:%M:%S")
    
    data = {
        "sensor": "온도",
        "value": random.randint(20, 30),
        "datetime": datetime.now().strftime("%Y.%m.%d %H:%M:%S")
    }
    
    # dict -> JSON 문자열
    str_json = json.dumps(data, ensure_ascii=False)
    
    return str_json
```

```python
temperature_sensor()
```

```python
# 온도 데이터를 파일에 저장하는 함수
import time


def save_templerture():
    file_path = os.path.join(os.getcwd(), "temperature.txt")
    with open(file_path, "at", encoding="UTF-8") as file:
        for i in range(5):
            data = temperature_sensor()
            print(data)
            file.write(data + "\n")
            time.sleep(1)

save_templerture()
```

```python
# 파일에서 온도 데이터를 읽어오는 함수
def read_temperature_from_file():
    file_path = os.path.join(os.getcwd(), "temperature.txt")
    with open(file_path, "rt", encoding="UTF-8") as file:
        while True:
            str_json = file.readline()
            if str_json == "":
                break
            # dict <-- json 문자열
            data = json.loads(str_json)
            print(data)
            
read_temperature_from_file()
```

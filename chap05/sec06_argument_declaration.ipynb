```python
def fun(*values):
    print(type(values))
    for i in range(len(values)):
        print(f"{i}: {values[i]}")
    
fun(1,2,3)
```

```python
def fun(a, b, *c, d, e=7):
    print(a, b, c, d, e)
    
fun(2, 3, d=4)
fun(1, 2, 3, 4, 5, d=6)
fun(1, 2, 3, 4, 5, e=6, d=7)
```

```python
def fun(*, a, b):
    print(a, b)

fun(a=1, b=2)
fun(b=1, a=2)
```

```python
def fun(a, b, *, c, d="Python"):
    print(a, b, c, d)
    
fun(1, 2, c=3)
# fun(1, 2, 3, 4)
fun(1, 2, d="Java", c=3)
```

```python
# / 앞: 위치 고정 매개변수
# / 뒤: 위치 또는 키워드 매개변수
def fun(a, b, /, c, d):
    print(a, b, c, d)

fun(1, 2, 3, 4)
fun(1, 2, c=3, d=4)
fun(1, 2, d=4, c=3)
```

```python
def fun(**a):
    print(type(a))
    for key in a:
        print(f"{key}: {a[key]}")
    
fun(name="파이썬", age=27)
```

```python
def fun(a, **b):
    print(a, b)
    
fun(1, k1=2, k2=3)
```

```python
person = ("홍길동", 30)
name, age = person
print(name, age)

person = ["홍길동", 30]
name, age = person
print(name, age)
```

```python
def fun(**args):
    print(args)

person = {"name":"홍길동", "age":30}

# 언패킹(key=value, ...)
fun(**person)
```

```python
def fun(*, mid, mname, mage):
    print(mid, mname, mage)
    
member = { "mid": "user1", "mname":"홍길동", "mage": 30}
fun(**member)
fun(mid="user1", mname="홍길동", mage=30)
```

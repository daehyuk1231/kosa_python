```python
# 함수 정의
def score_avg(scores: list[int]) -> float:
    sum: int = 0
    
    
    for i in range(len(scores)):
        sum += scores[i]
        
    avg: float = sum / len(scores)
    return avg

# 함수 호출
scores: list[int] = [90, 80, 70]
result: float = score_avg(scores)
print(result)
```

```python
if score_avg([90, 80, 70]) > 85:
    print("평균 점수가 85 이상입니다.")
else:
    print("평균값이 85 미만입니다.")
```

```python
def start() -> None:
    print("전원을 넣습니다.")
    print("로봇에 불이 들어옵니다.")
```

```python
def move() -> None:
    print("로봇이 움직입니다.")
    print("로봇이 저장된 위치에 도착합니다.")
```

```python
def stop() -> None:
    print("전원을 끕니다.")
    print("로봇에 불이 꺼집니다.")
```

```python
while True:
    print("명령어 입력:", end="")
    command: str = input()
    if command == "1":
        start()
    elif command == "2":
        move()
    elif command == "3":
        stop()
    else:
        print("처리할 수 없는 명령어 입니다.")
```

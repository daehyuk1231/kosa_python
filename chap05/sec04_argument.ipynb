```python
boards: list[dict] = []
```

```python
from datetime import datetime


def add_board(
    title: str,
    content: str,
    writer: str,
    # 기본값을 갖는 매개변수는 맨 마지막에 선언
    date: datetime = datetime.now(),
    hitcount: int = 0
) -> None:
    board: dict = {
        "title": title,
        "content": content,
        "writer": writer,
        "date": date,
        "hitcount": hitcount
    }
    boards.append(board)
```

```python
add_board("오늘은 비", "날씨가 춥네요", "user1")
add_board("오늘은 비", "날씨가 춥네요", "user1", datetime.now(), 5)
```

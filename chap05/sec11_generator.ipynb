```python
from typing import Generator


# Generator 생성 함수 정의
def create_generator() -> Generator[int, None, None]:
    # 값 생성
    value = 100
    while True:
        # 외부에 값 제공
        print("1) Generator에서 값 제공")
        print("2) Generator가 값 처리 완료될때까지 대기함")
        yield value
        print("4) Generator가 대기에서 풀림")
        value += 1
```

```python
# while문으로 외부에서 Generator 사용
# Generator 생성
generator = create_generator()
while True:
    # Generator 값 하나 받아오기
    value = next(generator)
    # 처리할때까지 5초가 걸린다고 가정
    time.sleep(5)
    print("3) 값 처리 완료")
```

```python
# for 외부에서 Generator 사용
# Generator 생성
generator = create_generator()
for value in generator:
    # Generator 값 하나 받아오기
    # 처리할때까지 5초가 걸린다고 가정
    time.sleep(5)
    print("3) 값 처리 완료")
```

```python
# 외부 처리값을 받아서 재가공해서 리턴하는 Generator 생성 함수 정의
# 첫번째 list: Generator가 제공하는 값의 타입
# 두번째 list: 외부에서 가공처리한 값을 받는 타입
# 세번째 str: Generator가 두번째에서 받은 값을 재가공해서 반환하는 값의 타입
def create_generator() -> Generator[list, list, str]:
    # 외부에 줄 list 생성
    source:list = [1, 2, 3]
    # 외부에 리스트를 제공
    print("1) 외부에 리스트를 제공")
    print("2) 외부에서 처리가 될때까지 대기")
    result:list = yield source
    # 재가공
    print("4) 외부 처리 결과를 재가공하기")
    sum = 0
    for value in result:
        sum += value
    final_result: str = f"Generator의 최종결과: {sum}"
    print("5) 재가공 결과를 반환")
    return final_result
```

```python
# 외부에서 Generator 사용하기
# Generator 생성
generator = create_generator()
# Generator로부터 값 가져오기
source: list = next(generator)
# 가져온 값을 처리
result: list = [ value*10 for value in source ]
# 처리된 결과를 다시 Generator로 보냄
try:
    time.sleep(5)
    print("3) 외부 처리 완료")
    generator.send(result)
except StopIteration as e:
    print(e.value)
```

```python
import random

# 제너레이터 생성 함수 정의
def create_generator() -> Generator[dict, None, None]:
    while True:
        # 센서로부터 현재 온도를 얻고 딕셔너리를 생성
        state: dict = {"sensor":"temperature", "value":random.randint(0, 30)}
        yield state

# 제너레이터를 사용
generator = create_generator()
for state in generator:
    print(state)
    time.sleep(5)
```

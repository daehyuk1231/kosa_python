```python
# 전역 네임스페이스에 선언된 변수(전역 변수)
var = 1

def fun():
    # 함수 네임스페이스에 선언된 변수(로컬 변수)
    var = 2
    print(var)
    print(var)
    
fun()
print(var)
```

```python
# 전역 네임스페이스에 선언된 변수(전역 변수)
var = 1

def fun():
    # 함수 네임스페이스에 선언된 변수(로컬 변수)
    var += 1
    print(var)
    
fun()
```

```python
# 전역 네임스페이스에 선언된 변수(전역 변수)
var = 1

def fun():
    global var
    var += 1
    print(var)
    
fun()
print(var)
```
